```python
import math
import torch
import gpytorch
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
from gpytorch.kernels import MaternKernel

sns.set_style('white')
# style.use('whitegrid')
palette = ["#1b4079", "#C6DDF0", "#048A81", "#B9E28C", "#8C2155", "#AF7595", "#E6480F", "#FA9500"]
sns.set(palette = palette, font_scale=2.0, style="white", rc={"lines.linewidth": 4.0})

%matplotlib inline
```

```python
torch.random.manual_seed(2021)

F0 = 100 ## init price
V0 = 0.25 ## init price
mu = 0.05 ## rate of return

alpha = 0.1
beta = 0.8
rho = -0.3 ## correlation between vol and forward
gamma = 0.5 ## correlation between two vols

n_task = 3

# T = 2 ## Time of Simulation
steps = 500 ## steps per time
dt = 1 / 252 ## delta t
```

```python
def cov_mat(n_task, rho, gamma):
    fwd_mat = torch.eye(n_task)
    fwd_vol_mat = torch.eye(n_task) * rho
    vol_mat = torch.ones(n_task, n_task)*gamma + torch.eye(n_task)*(1. - gamma)

    upper = torch.cat((fwd_mat, fwd_vol_mat), -1)
    lower = torch.cat((fwd_vol_mat.t(), vol_mat), -1)
    cov_mat = torch.cat((upper, lower), 0) 
    
    return cov_mat
```

```python
covar = cov_mat(n_task, rho, gamma)
```

```python
L = torch.cholesky(covar)
```

```python
DZ = L.matmul(torch.randn(n_task*2, steps) * np.sqrt(dt))

F = torch.zeros(steps, n_task)
V = torch.zeros(steps, n_task)
time = torch.linspace(0, steps * dt, steps)

F[0] = F0
V[0] = V0

for t in range(1, steps):
    F[t] = F[t-1] + V[t-1] * (F[t-1])**beta * DZ[:n_task, t]
    V[t] = V[t-1] + alpha * V[t-1]*DZ[n_task:, t]
```

```python
n_train = 350

train_x = time[:n_train]
test_x = time[n_train:]

# log_returns = torch.log(F.log()[1:]/F.log()[:-1]) 
log_returns = torch.log(F[1:]) - torch.log(F[:-1])

train_y, test_y = log_returns[:n_train], log_returns[n_train:]
```

## Define a new multi-task GP

We're defining an ICM variational GP here.


```python
from voltron.likelihoods import VolatilityGaussianLikelihood
from voltron.kernels import BMKernel
```

```python
from gpytorch.models import GP
from gpytorch.kernels import IndexKernel
from gpytorch.lazy import KroneckerProductLazyTensor
from gpytorch.distributions import MultitaskMultivariateNormal

class MultitaskVarGP(GP):
    def __init__(self, inducing_points, num_tasks, covar_module = None, rank = 1, **kwargs):
        super().__init__()
        
        self.register_parameter(
            "variational_mean", torch.nn.Parameter(
                torch.randn(inducing_points.shape[-1], num_tasks), requires_grad = True
            )
        )
        self.register_parameter(
            "variational_covar_root", torch.nn.Parameter(
                torch.eye(inducing_points.shape[-1]), requires_grad = True
            )
        )
        
        self.index_kernel = IndexKernel(num_tasks=num_tasks, rank=rank, **kwargs)
        self.data_kernel = covar_module
        self.inducing_points = inducing_points
        self.num_tasks = num_tasks
        
        ### we assume that the variational mean is N_ind x num_tasks and is free-form
        ### we assume that the variatioanl covariance is LL' \otimes I
        ### the identity comes from some really rough taylor expansion i did that suggests
        ### that it ought to cancel b/c we assume that the noise is independent across tasks
        
        
    @property
    def variational_strategy(self):
        # hacky af for now
        return self
    
    def kl_divergence(self):
        ## this computes KL(q || p)
        
        prior_dist = MultitaskMultivariateNormal(
            torch.zeros_like(self.variational_mean),
            KroneckerProductLazyTensor(self.data_kernel(self.inducing_points), self.index_kernel.covar_matrix)
        )
        var_dist = MultitaskMultivariateNormal(
            self.variational_mean,
            KroneckerProductLazyTensor(
                self.variational_covar_root @ self.variational_covar_root.t(),
                torch.eye(self.num_tasks)
            )
        )
        return torch.distributions.kl_divergence(var_dist, prior_dist)
    
    def forward(self, x, **kwargs):
        kuu = self.data_kernel(self.inducing_points)
        kux = self.data_kernel(self.inducing_points, x)
        kuu_inv_kux = kuu.inv_matmul(kux.evaluate())
        
        ### (Kxu Kuu^{-1} \otimes I)m
        mean_term = KroneckerProductLazyTensor(
            kuu_inv_kux.transpose(-1, -2), torch.eye(self.num_tasks)
        ).matmul(self.variational_mean.reshape(-1)).reshape(-1, self.num_tasks)
        
        ### covar is a sum of 3 terms
        ## T1 = Kxx \otimes I
        kxx = self.data_kernel(x)
        # first_covar_term = KroneckerProductLazyTensor(kxx, self.index_kernel.covar_matrix)
        
        ## T2 = - (Kxu Kuu^{-1} Kux \otimes Ktt)
        data_onto_inducing = kux.transpose(-1, -2).matmul(kuu_inv_kux)
        # second_covar_term = KroneckerProductLazyTensor(data_onto_inducing, self.index_kernel.covar_matrix).mul(-1.)

        first_covar_term = KroneckerProductLazyTensor(kxx - data_onto_inducing, self.index_kernel.covar_matrix)
        
        ## T3 = (Kxu Kuu^{-1} S Kuu^{-1} Kux \otimes I)
        variational_covar = self.variational_covar_root.matmul(self.variational_covar_root.T)
        data_onto_var_covar = kuu_inv_kux.transpose(-1, -2).matmul(
            variational_covar.matmul(kuu_inv_kux)
        )
        third_covar_term = KroneckerProductLazyTensor(data_onto_var_covar, torch.eye(self.num_tasks))
        
        ### Sigma = T1 + T2 + T3
        total_covar = first_covar_term + third_covar_term
        
        return MultitaskMultivariateNormal(mean_term, total_covar)
        
```

```python
from gpytorch.priors import LKJCovariancePrior, SmoothedBoxPrior

prior = LKJCovariancePrior(3, eta = 2.0, sd_prior=SmoothedBoxPrior(0.25, 2.0))

model = MultitaskVarGP(train_x, 3, BMKernel(), prior=prior)
```

```python
from gpytorch.mlls import VariationalELBO
```

```python
likelihood = VolatilityGaussianLikelihood(batch_shape = torch.Size((3,)), K = 1)
```

```python
mll = VariationalELBO(model = model, likelihood = likelihood, num_data = train_x.shape[0])
```

```python
%pdb
```

```python
# this is for running the notebook in our testing framework
import os
smoke_test = ('CI' in os.environ)
training_iterations = 2 if smoke_test else 250


# Find optimal model hyperparameters
model.train()
likelihood.train()

# Use the adam optimizer
params = list(model.parameters()) + list(likelihood.parameters())
optimizer = torch.optim.Adam(params, lr=0.1)
```

```python
for i in range(training_iterations):
    # Zero backpropped gradients from previous iteration
    optimizer.zero_grad()
    # Get predictive output
    with gpytorch.settings.num_gauss_hermite_locs(25), gpytorch.settings.max_cholesky_size(5000):       
        output = model(train_x)
        loss = -mll(output, train_y)
        loss.backward()
        if i % 50 == 0:
            print('Iter %d/%d - Loss: %.3f' % (i + 1, training_iterations, loss.item()))
            plt.show()
        optimizer.step()
```
