```python
from botorch.models import SingleTaskGP
import torch
from gpytorch.likelihoods import GaussianLikelihood
```

```python
time = torch.linspace(0, 1, 100)
vol = torch.sin(10. * time)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(time, vol)
```

```python
model = SingleTaskGP(time.view(-1,1), vol.view(-1,1))
model.likelihood.noise = 1e-4
```

```python
time_t = torch.linspace(1, 2, 100)
```

```python
post_over_vprime = model.posterior(time_t.view(-1,1))
```

```python
from gpytorch.kernels import Kernel

class VolatilityKernel(Kernel):
    has_lengthscale = False
    
    def __init__(self, num_locs=50, t_max=3):
        super().__init__()
        self.num_locs = num_locs
        self.register_parameter("latent_f", torch.nn.Parameter(torch.randn(self.num_locs)))
        self.bounds = torch.arange(0, t_max, t_max / self.num_locs)
        
    def forward(self, x1, x2, diag=False, **params):
        last_dim_is_batch = params.get("last_dim_is_batch", False)
        x1_ = x1.unsqueeze(-1)
        int_bounds = torch.minimum(x1_, x2).squeeze(-1)
        
        x1_ = x1
        x2_ = x1 if x2 is None else x2
        if last_dim_is_batch:
            x1_ = x1_.transpose(-1, -2).unsqueeze(-1)
            x2_ = x2_.transpose(-1, -2).unsqueeze(-1)
            
        x1_ = x1_.unsqueeze(-1)
        
        int_bounds = torch.minimum(x1_, x2_)
        int_values = ((int_bounds >= self.bounds) * self.latent_f.pow(2.0))
        int_x = self.bounds.expand_as(int_values)
        
        res = torch.trapz(y=int_values, x=int_x)
        if diag:
            return torch.diagonal(res, dim1=-2, dim2=-1)
        else:
            return res
```

```python
kernel = VolatilityKernel()
```

```python
kernel.bounds = torch.cat((time, time_t), dim=0)
kernel.latent_f.data = torch.cat((vol, post_over_vprime.rsample()[0,:,0]), dim=0)
```

```python
plt.plot(kernel.bounds, kernel.latent_f.data)
```

```python
joint_kernel = kernel(kernel.bounds).evaluate()
```

```python
plt.imshow(joint_kernel.detach())
```

```python
k_train = joint_kernel[:100, :100]
k_cross = joint_kernel[:100, 100:]
k_test = joint_kernel[100:, 100:]
```

```python
from gpytorch import lazify
```

```python
train_y = torch.sin(20 * time).view(-1,1).exp()
```

```python
kinv_y = lazify(k_train).add_jitter(1e-4).inv_matmul(train_y.log())
pred_mean = k_cross.matmul(kinv_y)
```

```python
plt.plot(time, train_y)
plt.plot(time_t, pred_mean.detach())
```
