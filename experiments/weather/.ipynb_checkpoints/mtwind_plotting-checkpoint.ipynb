```python
ntrain = 252
ntest = 126
neval = 25
```

```python
    import pickle as pkl
    import numpy as np
    
    stn_names, stn_lonlat, stn_data = pkl.load(open("./wind_data.p", 'rb'))
    lonlat = np.array(list(stn_lonlat.values()))
    full_dat = np.stack(list(stn_data.values()), -1).T
    full_dat[full_dat == -99.0] = 0.
    names = np.array(list(stn_names.values()))

    conus_idx = np.where(lonlat[:, 0] > -128)[0]
    lonlat = lonlat[conus_idx]
    full_dat = full_dat[conus_idx]
    names = names[conus_idx]

    full_data = torch.tensor(full_dat).float() + 1 # wind speeds can be zero.
    ndata = full_data.shape[-1]
    full_returns = torch.log(full_data[..., 1:] / full_data[..., :-1])

    T = 52
    ts = torch.linspace(0, T, ndata) + 1

    device = torch.device("cpu")

    # train on one full year worht of data and test on 6 mos for defaults
    # we can do logner
    split_diff = int((ndata - ntrain) / neval)
    train_starts = list(range(0, ndata - ntrain, split_diff))
    train_splits = [[x, x + ntrain] for x in train_starts]
    if train_splits[-1][-1] > (ndata-1):
        train_splits.pop(-1)
    eval_splits = [min(ndata-1, x + ntest + ntrain) for x in train_starts]
    print(train_splits, eval_splits)

    ts = ts.to(device)
    full_returns = full_returns.to(device)
    full_data = full_data.to(device)
```

```python
i = 0
```

```python
train_start, train_end = train_splits[i][0], train_splits[i][1]
test_end = eval_splits[i]

# filter out any tses w/ only missing data
keep_idx = np.where((full_data[:, train_start:train_end]-1).mean(-1).cpu() > 0.)[0]
print(keep_idx)
clonlat = lonlat[keep_idx]
full_data = full_data[keep_idx]
names = names[keep_idx]
```
